# Univariate analysis

In this notebook we will make use of the dataset that we clean. We will explore each one of the variable independently in order to get a deeper understanding of each one of them.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd

%matplotlib inline
```

Let's load our dataser

```python
df = pd.read_csv("data/2_prepared/clean_diabetes.csv")
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,7));
plt.hist(df['Glucose'].values, label="Glucose")
plt.hist(df['BMI'].values, label="BMI")
plt.hist(df['Age'].values, label="Age")
plt.legend()
plt.ylabel("Counts")

```

```python
d1 = df['Glucose'].values
d2 = df['BMI'].values
d3 = df['Age'].values

plt.figure(figsize=(15,7));

bins = np.linspace(min(d1.min(), d2.min(), d3.min()), max(d1.max(), d2.max(), d3.max()), 100)
counts1, _, _ = plt.hist(d1, bins=bins, label="Glucose")
plt.hist(d3, bins=bins, label="Age")
plt.hist(d2, bins=bins, label="BMI")

plt.legend()

plt.ylabel("Counts");
```

```python
plt.figure(figsize=(15,7));

bins = np.linspace(min(d1.min(), d2.min(), d3.min()), max(d1.max(), d2.max(), d3.max()), 100)
counts1, _, _ = plt.hist(d1, bins=bins, label="Glucose", density=True)
plt.hist(d2, bins=bins, label="BMI", density=True)
plt.hist(d3, bins=bins, label="Age", density=True)


plt.legend()

plt.ylabel("Probability");
```

```python
bins=50
plt.figure(figsize=(15,7));
plt.hist([d1, d2, d3], bins=bins, label="Stacked", density=True, histtype="barstacked", alpha=0.3)
plt.hist(d1, bins=bins, label="Glucose", density=True, histtype="step", lw=1)
plt.hist(d2, bins=bins, label="BMI", density=True, histtype="step", lw=1)
plt.hist(d3, bins=bins, label="Age", density=True, histtype="step", lw=1)
plt.legend()
plt.ylabel("Probability");
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
sb.boxplot(x="Outcome", y="Glucose", data=df, whis=3.0);
sb.swarmplot(x="Outcome", y="Glucose", data=df, size=5, ax=ax, color="k", alpha=0.3);

```

```python
fig, ax = plt.subplots(figsize=(15, 7))
# sb.violinplot(x="Outcome", y="Glucose", data=df, bw=0.2)
sb.violinplot(x="Outcome", y="Glucose", data=df, inner="quartile", bw=0.2);
sb.swarmplot(x="Outcome", y="Glucose", data=df, size=5, ax=ax, color="k", alpha=0.1);
```

```python
sd1 = np.sort(d1)
sd2 = np.sort(d2)
cdf = np.linspace(1/d1.size, 1, d1.size)

plt.plot(sd1, cdf, label="D1 CDF")
plt.plot(sd2, cdf, label="D2 CDF")
plt.hist(d1, histtype="step", density=True, alpha=0.3)
plt.hist(d2, histtype="step", density=True, alpha=0.3)
plt.legend();
```
